# Pie Chart Of Content Rating:

```python
df['rating'].value_counts().plot(kind='pie', autopct='%1.2f%%', startangle=960,labeldistance=1.2)
plt.title('Distribution of Content Ratings')
plt.ylabel('')
plt.show()
```

```python
sizes=df['rating'].value_counts()
labels=sizes.index
explode=[0.05]*len(sizes)
plt.figure(figsize=(10,6))
plt.pie(sizes,labels=labels,autopct='%1.2f%%',startangle=960,explode=explode,labeldistance=1.3,pctdistance=0.8)
plt.title('Distribution of Content Ratings',fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
sizes=df['rating'].value_counts()
labels=sizes.index
values=sizes.values

fig,ax=plt.subplots(figsize=(12, 10))
wedges, texts, autotexts = ax.pie(
    values,
    startangle=90,
    autopct='%1.1f%%',
    pctdistance=0.75,
    textprops=dict(color="w")
)

for i, p in enumerate(wedges):
    ang=(p.theta2 - p.theta1)/2. + p.theta1
    y=np.sin(np.deg2rad(ang))
    x=np.cos(np.deg2rad(ang))

    horizontal_alignment={-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle="angle,angleA=0,angleB={}".format(ang)

    ax.annotate(
        labels[i],
        xy=(x, y),
        xytext=(1.2*np.sign(x), 1.2*y),
        horizontalalignment=horizontal_alignment,
        arrowprops=dict(arrowstyle="-", connectionstyle=connectionstyle)
    )
ax.set_title('Distribution of Content Ratings', fontsize=16)
plt.tight_layout()
plt.show()
```

# Line Graph Of Titles Add Each Year:

```python
df['date_added']=pd.to_datetime(df['date_added'], errors='coerce')
df=df.dropna(subset=['date_added'])
df['year_added']=df['date_added'].dt.year
titles_per_year=df['year_added'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
titles_per_year.plot(kind='line', marker='o')
plt.title('Titles Added Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```
